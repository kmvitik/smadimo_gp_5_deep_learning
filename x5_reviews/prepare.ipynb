```python
import pandas as pd
```

```python
df = pd.read_csv("labled.csv")
df.head(3)
```

```python
df.drop(columns=["Unnamed: 0", "address", "rubrics", "is_grocery"], inplace=True)
df.head(3)
```

```python
categories = [
    "product_quality",
    "staff_service",
    "queues_checkout",
    "assortment_availability",
    "store_cleanliness",
    "other_problem",
]
for category in categories:
    v = df[category].value_counts()
    f = v.loc[False]
    t = v.loc[True]
    print(f"Category: {category}, count: {t}, percent: {t/(f+t)}")
```

```python
pd.set_option("display.max_colwidth", 300)
df[df["other_problem"]]["text"]
```

```python
main_categories = categories.copy()
main_categories.pop(-1)
```

```python
other_only = df[df["other_problem"] & ~df[main_categories].any(axis=1)]
len(other_only)
```

```python
other_only["name_ru"].value_counts()
```

```python
df["other_problem"] = df["other_problem"] & ~df[main_categories].any(axis=1)
df.head(10)
```

```python
for category in categories:
    v = df[category].value_counts()
    f = v.loc[False]
    t = v.loc[True]
    print(f"Category: {category}, count: {t}, percent: {t/(f+t)}")
```

```python
df["n_labels"] = df[categories].sum(axis=1)
print(df["n_labels"].value_counts())
df["n_labels"].describe()
```

```python
df.head(5)
```

```python
df.drop(columns=["name_ru", "rating", "len", "n_labels"], inplace=True)
df.head(3)
```

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42, shuffle=True)

val_df, test_df = train_test_split(
    temp_df, test_size=0.5, random_state=42, shuffle=True
)

train_df.to_csv("train.csv", index=False)
test_df.to_csv("test.csv", index=False)
val_df.to_csv("valid.csv", index=False)
```
