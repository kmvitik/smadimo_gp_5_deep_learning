```python
import pandas as pd
```

Работа с датасетом [geo-reviews-dataset-2023](https://huggingface.co/datasets/d0rj/geo-reviews-dataset-2023)

```python
df = pd.read_parquet(
    "hf://datasets/d0rj/geo-reviews-dataset-2023/data/train-00000-of-00001-49261e4e5a35a5a0.parquet"
)
df.head(3)
```

```python
print(df["name_ru"].value_counts())
x5 = df.loc[df["name_ru"] == "Пятёрочка"].copy()
```

```python
x5.loc[:, "city"] = x5["address"].str.split(",").str[0].str.strip()
print(x5["city"].value_counts().head(5))
print(x5["rating"].value_counts())
```

```python
x5.to_csv("x5.csv", index=False)
```

```python
df["is_grocery"] = df["rubrics"].apply(lambda x: "Магазин продуктов" in x)
print(df["is_grocery"].value_counts())
grocery_all = df[df["is_grocery"] == True]
```

```python
grocery_all["len"] = grocery_all["text"].str.len()
grocery_all["len"].describe()
```

```python
len(grocery_all[grocery_all["rating"] <= 2])
```

```python
grocery_all.to_csv("grocery_all.csv", index=False)
```
